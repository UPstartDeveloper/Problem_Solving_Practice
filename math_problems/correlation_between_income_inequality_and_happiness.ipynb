```
# Here is code to pull the dataset and relevant libraries 
# to help get you started - written by Eric Rood
import pandas as pd
gini_index = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/gini_index.csv')
happiness_index = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/happiness_index.csv')

gini_index.head()
```

```
happiness_index.head()
```

#### First, merge the two DataFrames

```
gini_index = gini_index.rename(columns={'country': 'country_or_region'})
merged = gini_index.merge(happiness_index, how='inner', on='country_or_region')
```

```
merged.head()
```

#### Let's Compute the Pearson Correlation Coefficient

```
import numpy as np 

def pearson_corr(x, y):
    '''Computes the Pearson correlation coefficient of 2 lists of numbers.'''
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    multiplied_deviations = [(i - x_mean)*(j - y_mean) for i,j in zip(x,y)]
    sq_deviation_1 = [(i - x_mean)**2 for i in x]
    sq_deviation_2 = [(j - y_mean)**2 for j in y]
    correlation_x_y = np.sum(
        multiplied_deviations) / \
        np.sqrt(np.sum(sq_deviation_1))/np.sqrt(np.sum(sq_deviation_2)
    )
    return correlation_x_y
```

```
income_ineq = merged["current"]
happiness = merged["overall_score"]
correlation_inequality_happines = pearson_corr(income_ineq, happiness)
```

```
print(f"Correlation between income inequality and happiness: {correlation_inequality_happines}.")
```
